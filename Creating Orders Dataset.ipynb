# Creating Customers Dataset

The objective is to join all tables of Olist dataset into one to carry out an exploration of the orders data relying on one single datasource.

```python
import pandas as pd
import numpy as np
import datetime
```

```python
customer_data=pd.read_csv('olist_customers_dataset.csv')
#orders_dataset contains dates, parse_dates arg should be called to convert them to datetime objects
order_data=pd.read_csv('olist_orders_dataset.csv',parse_dates=['order_purchase_timestamp','order_approved_at','order_delivered_carrier_date','order_delivered_customer_date','order_estimated_delivery_date'])
order_items=pd.read_csv('olist_order_items_dataset.csv')
payments=pd.read_csv('olist_order_payments_dataset.csv')
reviews=pd.read_csv('olist_order_reviews_dataset.csv')
products=pd.read_csv('olist_products_dataset.csv')
seller_geo=pd.read_csv('olist_sellers_dataset.csv')
product_category=pd.read_csv('product_category_name_translation.csv')
```

```python
# order_data: check if dates are datetime 
order_data.info()
```

```python
#order_data: check format of dates
order_data.head()
```

```python
# 1. merge customers and orders to merge customer data and order id, to later merge other tables
df=customer_data.merge(order_data, on='customer_id')
df.head()
```

```python
# 2. merge df with order items data for additional data related to each order
df=df.merge(order_items, on='order_id')
df=df[['customer_id', 'customer_unique_id', 'customer_zip_code_prefix',
       'customer_city', 'customer_state', 'order_id', 'order_status',
       'order_purchase_timestamp', 'order_approved_at',
       'order_delivered_carrier_date', 'order_delivered_customer_date',
       'order_estimated_delivery_date','product_id', 'seller_id', 'price',
       'freight_value']]
```

```python
# 3. merge df wtih payments
df=df.merge(payments,on='order_id')
df.head()
```

```python
# 4. merge df with reviews, left merge as not all reviews data have customer_id
df=df.merge(reviews, on='order_id')
df=df[['customer_id', 'customer_unique_id', 'customer_zip_code_prefix',
       'customer_city', 'customer_state', 'order_id', 'order_status',
       'order_purchase_timestamp', 'order_approved_at',
       'order_delivered_carrier_date', 'order_delivered_customer_date',
       'order_estimated_delivery_date', 'product_id',
       'seller_id', 'price', 'freight_value', 'payment_sequential',
       'payment_type', 'payment_installments', 'payment_value',
       'review_score']]
df.head()
```

```python
# 5. merge with products
df=df.merge(products, on='product_id')
df=df[['customer_id', 'customer_unique_id', 'customer_zip_code_prefix',
       'customer_city', 'customer_state', 'order_id', 'order_status',
       'order_purchase_timestamp', 'order_approved_at',
       'order_delivered_carrier_date', 'order_delivered_customer_date',
       'order_estimated_delivery_date', 'seller_id', 'price', 'freight_value', 'payment_sequential',
       'payment_type', 'payment_installments', 'payment_value', 'review_score',
       'product_category_name']]
df.head()
```

```python
# 6. merge with product_category to have prod cat english
df=df.merge(product_category, on='product_category_name')
#keep only englsh prod cat
df=df[['customer_id', 'customer_unique_id', 'customer_zip_code_prefix',
       'customer_city', 'customer_state', 'order_id', 'order_status',
       'order_purchase_timestamp', 'order_approved_at',
       'order_delivered_carrier_date', 'order_delivered_customer_date',
       'order_estimated_delivery_date', 'seller_id', 'price', 'freight_value', 'payment_sequential',
       'payment_type', 'payment_installments', 'payment_value', 'review_score','product_category_name_english']]
df.head()
```

```python
# 7. merge df with seller geo data
df=df.merge(seller_geo, on='seller_id')
df.head()
```

## Feature Engineering

- approval_timespan
- delivery_timespan
- delivered_on_time

```python
df['approval_timespan']=df['order_approved_at']-df['order_purchase_timestamp']
df.head()
```

```python
df['delivery_timespan']=df['order_delivered_customer_date']-df['order_approved_at']
df.head()
```

```python
def order_on_time(estimated_date, delivery_date):
    if estimated_date > delivery_date:
        value = 'Yes'
    else:
        value = 'No'
    return value
```

```python
df['order_on_time'] = df.apply(lambda x: order_on_time(x['order_estimated_delivery_date'], x['order_delivered_customer_date']), axis=1)
df.head()
```
